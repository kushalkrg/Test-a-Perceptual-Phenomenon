### Analyzing the Stroop Effect

(1) What is the independent variable? What is the dependent variable?

Independent variable: Color of the word

Dependent variable: Time of response

(2) What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

NULL Hypotheses($H_0$): 
In Null hypotheses we are going to challenge the stroops effect and prove that the congruent words condition, and an incongruent words condition donâ€™t differ in the response time.

$$H_0:\mu_d=0$$

Alternate Hypotheses($H_1$):
Alternate Hypotheses will be that stroops effect is indeed correct and there is a significant difference between response time for congruent words condition, and an incongruent words condition.

$$H_0:\mu_d\ne0$$

Where  :

$\mu_d$ :  true mean difference.

 the difference between paired values from two data sets.

$$d = x_1 - x_2$$

where $x_1$ is the value of variable x in the first data set, and $x_2$ is the value of the variable from the second data set that is paired with $x_1$.

The Statistical test we are going to use is the T-test.
We cannot use Z- test because it is used when we want to know whether our sample comes from a particular population, which is not or requirement.

Also in general we use t-test when:

    1.Sample has size below 30.

    2.Sample has an unknown population standard deviation.

Now the question is which type of t-test we should use ?

We will be using 'paired t-test' since this test shows the difference between two sample data sets.

We use the t-test for when we are studying two groups like in our case, response time of Congrunet and Incongruent tasks.

The paired t-test calculates the difference within each congruent-incongruent task's pair of response time, determines the mean of these changes, and reports whether this mean of the differences is statistically significant.

We want to determine performance of same subject on two different tasks,thus paired t-test.

Assumptions:

    1.Paired samples

    2.The differences of the pairs follow a normal distribution 

(3) Report some descriptive statistics regarding this dataset. Include at least one measure of central tendency and at least one measure of variability. The name of the data file is 'stroopdata.csv'.

```python
# Import packages
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline 

# Perform the analysis here

data = pd.read_csv('stroopdata.csv')

Congruent = data['Congruent'].values.tolist()
Incongruent = data['Incongruent'].values.tolist()

data.mean().plot('bar')
```

```python
print('Range of Congruent condition ' + str(max(Congruent)-min(Congruent)))
print('Range of Incongruent condition ' + str(max(Incongruent)-min(Incongruent)))
```

The mean response time for Incongruent condition is higher than that of Congruent condition

(4) Provide one or two visualizations that show the distribution of the sample data. Write one or two sentences noting what you observe about the plot or plots.

```python
# Build the visualizations here
fig, axs = plt.subplots(ncols=2,figsize=(15, 8))

length = len(Congruent)
width=0.30

plt1=axs[0].bar(range(length),Congruent,width,color='c')
plt2=axs[0].bar(map(lambda x:x+width,range(length)),Incongruent,width,color='r')


plt3=axs[1].plot(range(length),Congruent,'o-',color='c')
plt4=axs[1].plot(range(length),Incongruent,'o-',color='r')

axs[0].set_title('Bar Chart')
axs[0].legend((plt1[0], plt2[0]), ('Congruent', 'Incongruent'))
axs[0].set_xlabel('Words under test')
axs[0].set_ylabel('Time of Response')

axs[1].set_title('Line Plot')
axs[1].legend((plt3[0], plt4[0]), ('Congruent', 'Incongruent'))
axs[1].set_xlabel('Words under test')
axs[1].set_ylabel('Time of Response')

plt.show()

#reference- https://matplotlib.org
```

In both the charts above, we can see that the Incongruent condition's time of response is significantly higher than the Congruent condition's time of response. 

(5) Now, perform the statistical test and report the results. What is the confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

```python
# Perform the statistical test here
# Perform t-test
stats.ttest_rel(Congruent,Incongruent)
```

Since the p-value is very small (less than 0.05) we can reject the null hypothesis and our challange of proving the stroop's effect false,fails.
